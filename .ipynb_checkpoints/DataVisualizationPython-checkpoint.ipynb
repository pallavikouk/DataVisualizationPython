### Step-3 : Creating Beautiful Visualizations

#### Line Chart

```python
# Single line chart
import matplotlib.pyplot as plt
from matplotlib import style
# setting style for graphs
style.use('Solarize_Light2')
plt.rcParams['figure.figsize'] = (20,10)

figure = df.loc['Haiti', years].plot(kind = 'line', color = 'r')
plt.title('Immigration from Haiti to Canada from 1980-2013',color = 'black',size=30)
plt.xlabel('Year',size=30,color = 'black')
plt.ylabel('Number of immigrants',size=30,color = 'black')
plt.xticks(color = 'black',size=22)
plt.yticks(color = 'black',size=22)
plt.savefig('Line_Chart.png')

plt.show()
```

```python
df.loc['Haiti', years]
```

```python
# View immigration from China and India
# Line plot is useful to display several dependent variables against one independent variable. However, 
#it is recommended that no more than 5-9 lines on a single graph; any more than that and it becomes difficult to interpret.
```

```python
# Multiple Line chart 
import matplotlib.pyplot as plt
from matplotlib import style
# setting style for graphs
style.use('ggplot')
plt.rcParams['figure.figsize'] = (20,10)

df_CI = df.loc[['China','India'],years]
df_CI = df_CI.transpose()
df_CI.plot(kind = 'line')


plt.title('Immigration from China and India',size=30,color = 'black')
plt.ylabel('Number of immigrants',size=30,color = 'black')
plt.xlabel('Year',size=30,color = 'black')
plt.xticks(color = 'black',size=22)
plt.yticks(color = 'black',size=22)
plt.savefig('Multiple_Line_Chart.png')
plt.legend(['China', 'India'],loc='upper right', title='The Legend',prop={'size': 19},fontsize = 14,labelcolor='k')
plt.show()
```

```python
# Vertical bar chart
# Do not change the style back to ggplot 
# delete the style.use('ggplot') line of code
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (14,7))
#top_15.plot(kind = 'bar')
plots = sns.barplot(x='Country', y='total', data=df_top_5_country_and_totals)
plt.title('\n Immigration from top five countries contributing most to immigration \n',size=22)
plt.ylabel('\n Number of immigrants \n',size=22)
plt.xlabel('\n Country \n',size=22)
plt.xticks(color='white',size=22)
plt.yticks(color='white',size=22)

for bar in plots.patches:
    plots.annotate(format(int(bar.get_height()), ','),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=22, xytext=(0, 9),
                   textcoords='offset points')
    
        
plt.show
```

#### Doughnut Chart

```python
# Doughnut Chart
import matplotlib.pyplot as plt
import mplcyberpunk
plt.style.use('cyberpunk')
 
top5_total_list=list(df_top_5_country_and_totals['total'])
top5_total_list
top5_country_list=list(df_top_5_country_and_totals['Country'])
top5_country_list

 
# colors
colors = ['#FF0000', '#0000FF', '#FFFF00',
          '#ADFF2F', '#FFA500']
Top5_countries_list=list(df_top_5.columns)
Top5_countries_list
# explosion
explode = (0.05, 0.05, 0.05, 0.05, 0.05)
 
# Pie Chart
plt.pie(top5_total_list, colors=colors, labels=top5_country_list,
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
 
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
 
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
 
# Adding Title of chart
plt.title('Top 5 Immigrants from countries',color = 'white')
 
# Displaying Chart
plt.show()

```
